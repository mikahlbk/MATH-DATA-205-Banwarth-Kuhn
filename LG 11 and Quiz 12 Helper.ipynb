### Compute Eigenvalues

```python
import numpy as np

#make the matrix
A = np.array([[3,1,0],
              [1,3,0],
              [0,0,4]])

#find eigenvalues and eigenvectors
eigVals, eigVecs = np.linalg.eig(A)
print("eigVecs returns eigenvectors and eigVals returns eigenvalues.\n")

print("In particular, eigVecs returns eigenvectors as columns. The column eigVecs[:,i] corresponds to the eigenvalue eigVals[i]\n")

print("Eigenvalues:", eigVals)

print("\nEigenvectors:", eigVecs)

print("\nIn other words:\n")
for i in range(len(eigVals)):
    print(f"Eigenvalue {i}: {eigVals[i]}")
    print(f"Eigenvector {i}: {eigVecs[:, i]}\n")
```
