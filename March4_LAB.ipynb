## Part 1️⃣ — Warm-Up: When Can We Multiply Matrices?

Matrix multiplication is possible when:

$$A_{m \times n} \cdot B_{n \times p} = C_{m \times p}$$

The **inner dimensions must match**:

- The number of columns in \( A \) must equal  
- The number of rows in \( B \)

👉 **If the inner dimensions do not match, matrix multiplication is not possible.**

---

### ✅ Example

- \( A \) is **2 × 3**  
- \( B \) is **3 × 2**

Since the inner dimensions (3 and 3) match:

$$
(2 \times 3) \cdot (3 \times 2) \rightarrow (2 \times 2)
$$

The result will be a **2 × 2 matrix**.

```python
import numpy as np

#Matrix A = (2x3)
A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

#Matrix B = (3x2)
B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

#Vector v = (3x1)
v = np.array([
    [5],
    [3],
    [2]
])
mat_vec_result = np.dot(A,v)
print("A(v) should be 2x1)")
print("Shape is",mat_vec_result.shape)
print(mat_vec_result)
mat_mat_result = np.dot(A, B)
print("AB should be 2x2)")
print("Shape is", mat_mat_result.shape)
print(mat_mat_result)
```
