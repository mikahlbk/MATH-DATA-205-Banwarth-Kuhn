**Okay back to linear algebra.**

In our quest to learn as much as we can about vectors, we are moving on to multiplying two vectors. 

And so far we learned about this new tool we're calling the **dot product**. 

If we have two vectors
$$\vec{a} = (a_1, a_2, \dots, a_n)~\text{and}~\vec{b} = (b_1, b_2, \dots, b_n),$$

we multiply them in the following way

$$
\vec{a} \cdot \vec{b}
= \sum_{i=1}^{n} a_i b_i
= a_1 b_1 + a_2 b_2 + \cdots + a_n b_n.$$

Let's look at a concrete example in two dimensions. 

Let $\vec{a} = (1,3)$ and $\vec{b} = (4,-1)$. 

Then 
$$\vec{a} \cdot \vec{b}
= \sum_{i=1}^{2} a_i b_i
= a_1 b_1 + a_2 b_2 = 1\cdot4 + 3\cdot-1 = 4 -3 = 1.$$

Something to notice is that we start with two **vectors**, and we end up with a **scalar**.

Like with most linear algebra tools, there are two parts: the **computation** (you should know how to compute the dot product of two vectors) and the **geometric meaning.**

We are going to spend some time developing the geometric meaning now. But first, I want to show you how to use Python to compute the dot product to save you some heartache.

```python
#define the vectors
#vector_name = [v_1,v_2]
a = [1, 3]
b = [4, -1]

#define a variable to hold the outcome
#dot_product = a_1b_1+ a_2b_2
dot_product = a[0]*b[0] + a[1]*b[1]#why python,why???

#print result
print("Manual dot product = ",dot_product)

#we can also use a built-in Python function
#to use functions we load libraries
import numpy as np
dot_product_func = np.dot(a, b)
print("Function dot product =",dot_product_func)
```

Yay, they match!
