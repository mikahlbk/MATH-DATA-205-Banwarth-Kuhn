```python
import numpy as np
# ----- How to make a VECTOR -----
u = np.array([1, 2]) 
v= np.array([.5, 3]) 

print("u =", u)
print("v =", v)
```

```python
# ----- How to compute the DOT PRODUCT -----
dot_prod = np.dot(u,v)      
print("u · v =", dot_prod)
```

```python
# ----- How to make a MATRIX -----
A = np.array([
    [1, 2],
    [3, 4]
])

print("A=", A)
```

```python
# ----- How to do Matrix Vector Multiplication -----
Au = np.dot(A, u)
Av = np.dot(A, v)

print("Au =", Au)
print("Av =", Av)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ----- How to Plot Vectors-----
plt.figure()

# origin point
origin = [0, 0]

# plot each vector using arrows
plt.quiver(*origin, *u, angles='xy', scale_units='xy', scale=1, color = 'b',label='u')
plt.quiver(*origin, *Au, angles='xy', scale_units='xy', scale=1, color = 'r',label='Au')
plt.quiver(*origin, *v, angles='xy', scale_units='xy', scale=1, color = 'g',label='v')
plt.quiver(*origin, *Av, angles='xy', scale_units='xy', scale=1, label='Av')

# ----- Formatting -----
plt.xlim(-10, 15)
plt.ylim(-10, 15)

plt.axhline(0)
plt.axvline(0)

plt.grid(True)
plt.legend()

plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```
