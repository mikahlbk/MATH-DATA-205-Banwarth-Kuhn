## Exercise 4.1.2

```python
import numpy as np

#make the matrix
A = np.array([[3,1,0],
              [1,3,0],
              [0,0,4]])

#find eigenvalues and eigenvectors
eigVals, eigVecs = np.linalg.eig(A)
print("eigVecs returns eigenvectors and eigVals returns eigenvalues.\n")

print("In particular, eigVecs returns eigenvectors as columns. The column eigVecs[:,i] corresponds to the eigenvalue eigVals[i]\n")

print("Eigenvalues:", eigVals)

print("\nEigenvectors:", eigVecs)

print("\nIn other words:\n")
for i in range(len(eigVals)):
    print(f"Eigenvalue {i}: {eigVals[i]}")
    print(f"Eigenvector {i}: {eigVecs[:, i]}\n")
```

```python
import numpy as np
#make the matrix P
P = np.array([[0.70710678,-0.70710678 ,0],
              [0.70710678,0.70710678,0],
             [0,0,1]])
#make the matrix D
D = np.array([[4 ,0,0],
              [0,2,0],
             [0,0,4]])
#find P inverse
print("Inverse of P:")
P_inv = np.linalg.inv(P)
print(P_inv)
#verify B1 = PDP^-1
step1_mat = np.dot(D,P_inv) 
step2_mat = np.dot(P,step1_mat)
print("PDP^-1 =")
print(step2_mat)
```

```python
import numpy as np
#make the matrix P
P = np.array([[0.89442719, -0.70710678 ],
              [0.4472136 ,0.70710678]])
#make the matrix D
D = np.array([[5 ,0],
              [0,2]])
#find P inverse
print("Inverse of P:")
P_inv = np.linalg.inv(P)
print(P_inv)
#verify B1 = PDP^-1
step1_mat = np.dot(D,P_inv) 
step2_mat = np.dot(P,step1_mat)
print("PDP^-1 =")
print(step2_mat)
```

```python
import numpy as np
#make the vector
v = np.array([1, 0]) 
#multiply matrix times vector
v1 = np.dot(P_inv,v)      
print("v1 = P_inv · v =", v1)

#multiply matrix times vector
v2 = np.dot(D,v1)      
print("v2 = D · v1 =", v2)

#Hmultiply matrix times vector
v3 = np.dot(P,v2)      
print("v3 = P · v2 =", v3)
```

```python
import numpy as np
magnitude = np.linalg.norm(v)
print("|v| =", magnitude)

magnitude = np.linalg.norm(v1)
print("|v1| =", magnitude)
```

```python
import numpy as np
j =  np.array([0, 1]) 
p_inv_j = np.dot(P_inv,j)      
dot_prod = np.dot(v1,p_inv_j)
print("dot product =",dot_prod)
```

```python
import numpy as np
# ----- How to make a MATRIX -----
A = np.array([[1,1],
              [0,1]])

print("A=", A)
```

```python
import numpy as np
# ----- How to make a VECTOR -----
i = np.array([1, 0]) 
j= np.array([0, 1]) 

print("i =", i)
print("j =", j)
```

```python
import numpy as np
# ----- How to do Matrix Vector Multiplication -----
Ai = np.dot(A, i)
Aj = np.dot(A, j)

print("Ai =", Ai)
print("Aj =", Aj)
```

```python
import numpy as np
#compute length of vecror
length = np.linalg.norm(Ai)
print("|Ai| =", length)
```

```python
import numpy as np
# ----- How to compute the DOT PRODUCT -----
dot_prod = np.dot(Ai,Aj)      
print("Ai · Aj =", dot_prod)
```
