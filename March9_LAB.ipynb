```python
import numpy as np
# ============================================================
# Part 1: Determinants
# ============================================================

print("=== PART 1: DETERMINANTS ===\n")

A = np.array([[2, 1],
              [5, 3]], dtype=float)

B = np.array([[1, 2],
              [2, 4]], dtype=float)   # singular matrix

C = np.array([[1, 2, 3],
              [0, 1, 4],
              [5, 6, 0]], dtype=float)

print("Matrix A:\n", A)
print("det(A) =", np.linalg.det(A))
print()

print("Matrix B:\n", B)
print("det(B) =", np.linalg.det(B))
print("Notice det(B) is 0 → This matrix is singular (not invertible)\n")

print("Matrix C:\n", C)
print("det(C) =", np.linalg.det(C))
print()
```
