## Change of Basis Example

We are given an alternative basis

$$ A = \{\vec{w} = (2,1), \vec{z}=(-1,2)\}$$


We want to write the vector

$$\vec{u} = (5,3)$$

as a linear combination of the alternative basis vectors, i.e.

$$\vec{u} = \alpha_A \vec{w} + \beta_A \vec{z}$$

This means we must solve:

$$ \alpha_A \begin{pmatrix}2 \\ 1 \end{pmatrix}
+
\beta_A \begin{pmatrix}-1 \\ 2 \end{pmatrix}
=
\begin{pmatrix}5 \\ 3 \end{pmatrix}.$$

This gives the system of equations:

$$
\begin{cases}
2\alpha_A - \beta_A = 5 \\
\alpha_A + 2\beta_A = 3
\end{cases}
$$

We now solve this system using Python.


```python
import sympy as sp

# Define unknowns
alpha_A, beta_A = sp.symbols('alpha_A beta_A')

# Define equations
eq1 = sp.Eq(2*alpha_A - beta_A, 5)
eq2 = sp.Eq(alpha_A + 2*beta_A, 3)

# Solve system
solution = sp.solve((eq1, eq2), (alpha_A, beta_A))

print(solution)
```

## Applying Change of Basis to Student Performance Data

## 1. Plot the data

```python
import matplotlib.pyplot as plt

# Data
students = ['A', 'B', 'C', 'D', 'E', 'F']
homework_scores = [65, 70, 75, 80, 85, 90]
exam_scores = [68, 72, 76, 82, 86, 91]

# Create scatter plot
plt.figure()
plt.scatter(homework_scores, exam_scores)

# Label each point with the student name
for i, student in enumerate(students):
    plt.text(homework_scores[i]+.5, exam_scores[i], student)

# Labels and title
plt.xlabel("Homework Score")
plt.ylabel("Exam Score")

# Show plot
plt.show()
```

## 2. Plot the alternative basis vectors on same plot as data points

```python
import matplotlib.pyplot as plt

# Data
students = ['A', 'B', 'C', 'D', 'E', 'F']
homework_scores = [65, 70, 75, 80, 85, 90]
exam_scores = [68, 72, 76, 82, 86, 91]

# Create figure
plt.figure()

# Scatter plot
plt.scatter(homework_scores, exam_scores)

# Label each point
for i, student in enumerate(students):
    plt.text(homework_scores[i], exam_scores[i]+5, student)

# ---- Add Dashed Vectors from (0,0) ----
scale = 95

# Vector (1,1)
plt.plot([0, scale], [0, scale], linestyle='--')
plt.text(scale, scale, r"$\vec{w}=(1,1)$")

# Vector (1,-1)
plt.plot([0, scale], [0, -scale], linestyle='--')
plt.text(scale, -scale, r"$\vec{z} = (1,-1)$")

# Axis settings
plt.xlim(60, 100)
plt.ylim(-100, 100)

# Move x-axis to y = 0
ax = plt.gca()
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Labels
plt.xlabel("Homework Score")
plt.ylabel("Exam Score")

plt.show()

```
